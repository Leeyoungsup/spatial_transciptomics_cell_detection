```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
```

```python
# Parquet 파일 읽기
# 파일 경로를 실제 parquet 파일 경로로 변경하세요
file_path = '../../data/spatialTranscriptome/cellvit_seg/NCBI783_cellvit_seg.parquet'

# parquet 파일 읽기
df = pd.read_parquet(file_path)

# 데이터프레임 정보 출력
print(f"데이터 shape: {df.shape}")
print(f"\n컬럼 목록:\n{df.columns.tolist()}")
print(f"\n데이터 타입:\n{df.dtypes}")
print(f"\n첫 5개 행:\n{df.head()}")
```

```python
file_list=glob('../../data/spatialTranscriptome/transcripts/TENX*.parquet')
RnA_unique_list=[]
for i in tqdm(range(len(file_list))):
    df_transcript = pd.read_parquet(file_list[i])
    df_transcript=df_transcript.loc[df_transcript['overlaps_nucleus']==1]
    if type(df_transcript['feature_name'].iloc[0]) == bytes:
        df_transcript['feature_name'] = df_transcript['feature_name'].str.decode('utf-8')
        df_transcript = df_transcript[~df_transcript['feature_name'].str.contains('BLANK|NegControl|antisense', case=False, na=False)]
    elif type(df_transcript['feature_name'].iloc[0]) == str:
        df_transcript = df_transcript[~df_transcript['feature_name'].str.contains('BLANK|NegControl|antisense', case=False, na=False)]

    RnA_unique_list.extend(list(pd.unique(df_transcript['feature_name'])))
    RnA_unique_list=list(np.unique(np.array(RnA_unique_list)))
```

```python
pd.DataFrame(RnA_unique_list).to_csv('../../data/spatialTranscriptome/ALL_TENX_transcripts.csv', index=False, header=False)
```
